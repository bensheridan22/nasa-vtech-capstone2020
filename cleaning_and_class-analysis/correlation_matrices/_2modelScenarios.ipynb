# Split by Class

```python
monte_carlo_alg2 = models_2[mc_alg]
approx_control_var_alg2 = models_2[acv_alg]
parametric_acv_alg2 = models_2[pacv_alg]

data1 = monte_carlo_alg2.iloc[:, 27:30].dropna()
data2 = approx_control_var_alg2.iloc[:, 27:32].dropna()
data3 = parametric_acv_alg2.iloc[:, 27:35].dropna()
```

```python
covMatrix1 = np.corrcoef(data1.T, bias=True)
cols1 = data1.columns

covMatrix2 = np.corrcoef(data2.T, bias=True)
cols2 = data2.columns

covMatrix3 = np.corrcoef(data3.T, bias=True)
cols3 = data3.columns
```

```python
cov_mat = covMatrix1
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(cov_mat,
                 cbar=True,
                 annot=True,
                 square=True,
                 cmap='coolwarm',  
                 fmt='.5f',
                 yticklabels=cols1,
                 xticklabels=cols1)
plt.title('Correlation Matrix of Analytical AMS Algorithms - 2 Model Scen', size = 20)
plt.tight_layout()
plt.savefig("/Users/bensheridan/Documents/Senior/CMDACapstone/figures/mod2AMSanalytical.jpg")
```

```python
cov_mat = covMatrix2
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(cov_mat,
                 cbar=True,
                 annot=True,
                 square=True,
                 cmap='coolwarm',   
                 fmt='.5f',
                 yticklabels=cols2,
                 xticklabels=cols2)
plt.title('Correlation Matrix of ACV Algorithms - 2 Model Scenario', size = 20)
plt.tight_layout()
plt.savefig("/Users/bensheridan/Documents/Senior/CMDACapstone/figures/mod2acv.jpg")
```
