# All algorithms

```python
all_data = models_4[["mlmc ams", 'mfmc ams', 'wrdiff ams', 'grdsr ams','grdmr ams', 'acvis ams', 'gissr ams','gismr ams', 'acvmf ams','acvkl ams','acvmfu ams', 'gmfsr ams','gmfmr ams']]
all_alg_data = all_data.dropna()
all_alg_corr_matrix = np.corrcoef(all_alg_data.T, bias=True)
cols = all_alg_data.columns

cov_mat = all_alg_corr_matrix
plt.figure(figsize=(15,15))
sns.set(font_scale=1.5)
hm = sns.heatmap(cov_mat,
                 cbar=True,
                 annot=True,
                 square=True,
                 cmap='coolwarm',
                 fmt='.3f',
                 yticklabels=cols,
                 xticklabels=cols)
plt.title('Correlation Matrix of All AMS Algorithms - 4 Model Scenario', size = 25)
plt.tight_layout()
plt.savefig("/Users/bensheridan/Documents/Senior/CMDACapstone/figures/mod4/mod4AMSall.jpg")
```
